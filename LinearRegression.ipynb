## How is the data structured?

The data consists of the historical quotes for the SP500 - Open, Close, High, Low, Volume and Adjust close prices and most importantly all indexed by timestamps. This is a Time series -  where the data is observed and measured against time and the individual details eg: High and Low price points, become the cross-section of the data.

Time series data have a natural temporal ordering. This makes time series analysis distinct from cross-sectional studies, in which there is no natural ordering of the observations (src:wikipedia)

The plot below shows the fluctuation of the SP500 closing prices for 2010 to 2014. There is an overlay of short term and long term trends.

```python
# Calculate the windowed trends

spc['MA42'] = pd.rolling_mean(spc['Open'], window = 42, min_periods = 30)
spc['MA252'] = pd.rolling_mean(spc['Open'], window = 252,  min_periods = 30)

spc[['Open', 'MA252', 'MA42']].plot()

plt.legend(["Actual", "Long term", "Short term"])
plt.title('Trends in the SP500 data', fontsize = 20)
#plt.savefig('./present/Trends.png', bbox_inches='tight')
```

##Preprocessing: What features do I use?

This is where I spent most of the time, because I had to generate the features necessary to run the regression model. The selected/generated features are:
+ Daily Log returns : Gives us a measure how much the stock changes from day to day. 
+ Trading Volume:  Trading volume indicates how actively is the stock traded and seems indicative of price's moving direction 
+ Short term Moving average: Smoothes out the daily fluctuations, while retaining the trend
+ Volatility: A measure of risk calculated using the weighted moving standard deviation of the returns. 

###A look at how the returns are structured. 
The histogram tells us that in that past 4 years we have had mostly small positive returns. 

```python
# Daily Log returns

spc['Returns'] = np.log(spc['Open'] / spc['Open'].shift(1))    
spc["Returns"].hist(bins = 50)
plt.title('Histogram of the Log Returns', fontsize = 20)

# For the presentation
#plt.savefig('./present/Returns_hist.png', bbox_inches='tight')
```

```python
# Calculate Volatility
spc['Mov_Vol'] = pd.rolling_std(spc['Returns'], window = 252,   min_periods = 30) * np.sqrt(252)
spc['Exp_Vol'] = pd.ewmstd(spc['Returns'], span = 250, bias = True) * np.sqrt(250)
spc[['Returns', 'Mov_Vol', 'Exp_Vol']].plot()

plt.legend(["Returns", "Moving Volatility", "Exp Weighted Volatility"])
plt.title('Volatility in the Returns ', fontsize = 20)
```

### Partitioning the data for a Training and Testing:
The test/train splitting cross-validation approach selects random points from the entire sample to create a training set and testing set. This works well for cross-sectional data but with time series data, we need a temporal split. Meaning, don't use future data to predict something now and avoid causal modeling. (src: Doing data science). 

There are 2 methods used: 
+ Simply partitioning the data based on its time index 
+ Accumulate a moving window of training data

```python
# Partition the data in time
train_start = dt.datetime(2012, 5, 1)
test_start = dt.datetime(2014, 5, 1)
```

```python
# Create training set
X_train = spc[['Exp_Vol', "Volume", "Returns", 'MA42']][(spc.index > train_start) & (spc.index < test_start)]
y_train = spc["Adj Close"][(spc.index > train_start) & (spc.index < test_start)]
```

```python
# Create testing set
X_test = spc[['Exp_Vol', "Volume", "Returns", 'MA42']][(spc.index >= test_start)]
y_test = spc["Adj Close"][(spc.index >= test_start)]

# Use to fix "Series" if only one feature is used, Linear regression requires a 2D input
#X_train = X_train.reshape(X_train.shape[0], 1)
#X_test = X_test.reshape(X_test.shape[0], 1)
```

##Predictive Modeling using Linear Regression
We are working with continuous data and I have some output already, how do I model my problem? Supervised Learning - Regression techniques to model a continuous but complicated relationship. Here I am using a training period of 2 years to predict the next 5 months. The model uses plain vanilla linear regression and ridge regression. One thing to note is that, even after testing the problem with varying time periods, linear regression always scored better on the training set. This is in contrast to ridge regression which always scored better on the testing set. I am not quite sure why, but a guess is that ridge regression might be game-day method.

```python
from sklearn.linear_model import LinearRegression, Ridge

regr = LinearRegression()
regr.fit(X_train, y_train)

print "LinearRegression Model Scores"
print 'Training score ', regr.score(X_train, y_train)
print 'Testing score ', regr.score(X_test, y_test)

ridge = Ridge(alpha = 0.05)
ridge.fit(X_train, y_train)

print "\nRidge Regression Model Scores"
print 'Training score ', ridge.score(X_train, y_train)
print 'Testing score ', ridge.score(X_test, y_test)
```

```python
y_test.plot(label = 'Actual Price Fluctuation',  linewidth = 3)
plt.plot(X_test.index, regr.predict(X_test), 'r-', label = 'Linear')
plt.plot(X_test.index, ridge.predict(X_test), 'm--', label = 'Ridge', linewidth = 2)

plt.legend(loc = 'best')
plt.title('Prediction Model Comparision', fontsize = 20)
#plt.savefig('./present/Model.png', bbox_inches='tight')
plt.show()
```

In the figure above, we see three line plots. The blue one is the actual closing prices, the red and dotted magenta are the predicted prices by the Linear and Ridge models respectively. While the model retains a similar shape, we can see that if there major fluctuations in the market the model is slow to catch up. The in-sample was very high, but the out-of-sample scores not so much. The same model does not work as well for the closing prices during late October and November

## Windowed Regression
Lets look at many small models instead of a single large model see if that improves the accuracy. The below experiment, uses a moving window of the training data and regression modelling. Setting up a feedback loop. Since we are looking at shorter time frames, I generated 2 newer features :
+ Moving Volatility: Non-weighted scheme, rolling standard deviation for a 30 day window
+ 10 day moving average : a trend for a much shorter period.

The loop compares linear regression with ridge regression and uses the model that scores better on the current training window.
The figure shows the plot of the actual price vs the predicted price for a 4 year period. The red band is a 3% confidence interval in the model.




```python
spc['Mov_Vol1'] = pd.rolling_std(spc['Returns'], window = 30) * np.sqrt(30)
spc['MA10'] = pd.rolling_mean(spc['Open'], window = 5)
```

```python
spc = spc.dropna()
X = spc[['Mov_Vol1', "Volume", "Returns", 'MA10']]
y = spc["Adj Close"]
```

```python
train_period = dt.timedelta(days = 180)
test_period = dt.timedelta(days = 30)

start_date = dt.datetime(2010, 1, 1)
end_date = dt.datetime(2014, 10, 1)
```

```python
# Apply windowed regression

regr_score, ridge_score = 0.0, 0.0

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Windowed Regression Model', fontsize = 20)


train_start = start_date
test_start = train_start + train_period
test_end = test_start + test_period 

curr_date = train_start


while curr_date < end_date:
    # create training set
    X_train = X[(X.index >= train_start) & (X.index < test_start)]
    y_train = y[(y.index >= train_start) & (y.index < test_start)]
    
    # create test set
    X_test = X[(X.index >= test_start) & (X.index <= test_end)]
    y_test = y[(y.index >= test_start) & (y.index <= test_end)]
    
    # set time parameters for the next window
    train_start += test_period 
    test_start = train_start + train_period
    test_end = test_start + test_period 
    curr_date = train_start
    
    # plot the test data
    ax.plot(y_test.index, y_test, 'b-')    
    
    # fit the training data using Linear regression
    regr = LinearRegression()
    regr.fit(X_train, y_train)
    regr_score = regr.score(X_train, y_train)   

    # fit the training data using Ridge regression
    ridge = Ridge(alpha = 0.01)
    ridge.fit(X_train, y_train)
    ridge_score = ridge.score(X_train, y_train) 
    
    # choose the better fit
    if regr_score > ridge_score:
        y_pred = regr.predict(X_test)
    else:
        y_pred = ridge.predict(X_test)
    
    # Create error vector to plot the Confidence interval
    error = y_pred * 0.03
    # Plot predicted prices & confidence interval
    ax.plot(y_test.index, y_pred, 'r--')
    ax.fill_between(y_test.index, (y_pred - error), (y_pred + error), alpha = 0.2, facecolor = 'red')


ax.legend(["Actual Price", "Pred Price"])
#plt.savefig('./present/Window.png', bbox_inches='tight')
```
