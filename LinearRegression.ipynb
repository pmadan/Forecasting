```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

pd.set_option('display.mpl_style', 'default')

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
import pandas.io.data as web
```

```python
spc = web.DataReader(name = "^GSPC", data_source = 'yahoo', start = '2010-1-1', end = '2014-10-1')
```

```python
# Daily Log returns

spc['Returns'] = np.log(spc['Adj Close'] / spc['Adj Close'].shift(1))    
spc["Returns"].plot()
```

```python
spc["Returns"].hist(bins = 50)
```

```python
spc['MA42'] = pd.rolling_mean(spc['Adj Close'], window = 42, min_periods = 30)
spc['MA252'] = pd.rolling_mean(spc['Adj Close'], window = 252,  min_periods = 30)
#spc['STD250'] = pd.rolling_std(spc['Adj Close'], 250, min_periods = 10)
spc[['Adj Close', 'MA252', 'MA42']].plot()
```

```python
import math
# Mov_Vol - Moving volatility
spc['Mov_Vol'] = pd.rolling_std(spc['Returns'], window = 252,   min_periods = 30) * math.sqrt(252)
spc['Exp_Vol'] = pd.ewmstd(spc['Returns'], span = 250) * np.sqrt(250)
spc[['Returns', 'Mov_Vol', 'Exp_Vol']].plot()
```

```python
train_start = dt.datetime(2012, 5, 1)
test_start = dt.datetime(2014, 5, 1)
```

```python
X_train = spc[['Exp_Vol', "Volume", "Returns", 'MA42']][(spc.index > train_start) & (spc.index < test_start)]
y_train = spc["Adj Close"][(spc.index > train_start) & (spc.index < test_start)]

# Use to fix "Series" if only one feature is used
#X_train = X_train.reshape(X_train.shape[0], 1)
#y_train = y_train.reshape(y_train.shape[0], 1)
```

```python
X_test = spc[['Exp_Vol', "Volume", "Returns", 'MA42']][(spc.index >= test_start)]
y_test = spc["Adj Close"][(spc.index >= test_start)]
#X_test = X_test.reshape(X_test.shape[0], 1)
#y_test = y_test.reshape(y_test.shape[0], 1)
```

```python
from sklearn.linear_model import LinearRegression, Ridge
regr = LinearRegression()
regr.fit(X_train, y_train)

print 'Training score ', regr.score(X_train, y_train)
print 'Testing score ', regr.score(X_test, y_test)
```

```python
y_test.plot()
plt.plot(X_test.index, regr.predict(X_test), color = 'red')
plt.show()
```
