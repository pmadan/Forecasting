```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

pd.set_option('display.mpl_style', 'default')

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
import pandas.io.data as web
```

```python
spc = web.DataReader(name = "^GSPC", data_source = 'yahoo', start = '2010-1-1', end = '2014-10-1')

#spc.to_csv("SP500.csv")
spc.head()
```

```python
# Daily Log returns

spc['Returns'] = np.log(spc['Open'] / spc['Open'].shift(1))    
spc["Returns"].hist(bins = 50)
plt.title('Histogram of the Log Returns', fontsize = 20)

# For the presentation
#plt.savefig('./present/Returns_hist.png', bbox_inches='tight')
```

```python
spc['MA42'] = pd.rolling_mean(spc['Open'], window = 42, min_periods = 30)
spc['MA252'] = pd.rolling_mean(spc['Open'], window = 252,  min_periods = 30)

spc[['Adj Close', 'MA252', 'MA42']].plot()

plt.legend(["Actual", "Long term", "Short term"])
plt.title('Trends in the SP500 data', fontsize = 20)
#plt.savefig('./present/Trends.png', bbox_inches='tight')
```

```python
spc['Mov_Vol'] = pd.rolling_std(spc['Returns'], window = 252,   min_periods = 30) * np.sqrt(252)
spc['Exp_Vol'] = pd.ewmstd(spc['Returns'], span = 250, bias = True) * np.sqrt(250)
spc[['Returns', 'Mov_Vol', 'Exp_Vol']].plot()

plt.legend(["Returns", "Moving Volatility", "Exp Weighted Volatility"])
plt.title('Volatility in the Returns ', fontsize = 20)
```

```python
train_start = dt.datetime(2012, 5, 1)
test_start = dt.datetime(2014, 5, 1)
```

```python
X_train = spc[['Exp_Vol', "Volume", "Returns", 'MA42']][(spc.index > train_start) & (spc.index < test_start)]
y_train = spc["Adj Close"][(spc.index > train_start) & (spc.index < test_start)]
print X_train.shape
print y_train.shape
```

```python
X_test = spc[['Exp_Vol', "Volume", "Returns", 'MA42']][(spc.index >= test_start)]
y_test = spc["Adj Close"][(spc.index >= test_start)]

# Use to fix "Series" if only one feature is used, Linear regression requires a 2D input
#X_train = X_train.reshape(X_train.shape[0], 1)
#X_test = X_test.reshape(X_test.shape[0], 1)
```

```python
from sklearn.linear_model import LinearRegression, Ridge

regr = LinearRegression()
regr.fit(X_train, y_train)

print "LinearRegression Model Scores"
print 'Training score ', regr.score(X_train, y_train)
print 'Testing score ', regr.score(X_test, y_test)

ridge = Ridge(alpha = 0.01)
ridge.fit(X_train, y_train)

print "\nRidge Regression Model Scores"
print 'Training score ', ridge.score(X_train, y_train)
print 'Testing score ', ridge.score(X_test, y_test)
```

```python
y_test.plot(label = 'Actual Price Fluctuation',  linewidth = 3)
plt.plot(X_test.index, regr.predict(X_test), 'r-', label = 'Linear')
plt.plot(X_test.index, ridge.predict(X_test), 'm--', label = 'Ridge', linewidth = 2)

plt.legend(loc = 'best')
plt.title('Prediction Model Comparision', fontsize = 20)
plt.show()
#plt.savefig('./present/Model.png', bbox_inches='tight')
```

```python
spc['Exp_Vol1'] = pd.ewmstd(spc['Returns'], span = 30, bias = True) * np.sqrt(30)
spc['MA10'] = pd.rolling_mean(spc['Open'], window = 10)
```

```python
spc = spc.dropna()
X = spc[['Exp_Vol1', "Volume", "Returns", 'MA10']]
y = spc["Adj Close"]
```

```python
start_date = dt.datetime(2010, 1, 1)
end_date = dt.datetime(2014, 10, 1)

train_period = dt.timedelta(days = 180)
test_period = dt.timedelta(days = 45)
```

```python


regr_score, ridge_score = 0.0, 0.0

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Windowed Regression Model', fontsize = 20)


train_start = start_date
test_start = train_start + train_period
test_end = test_start + test_period 

curr_date = train_start

y_pred = []
while (curr_date + train_period) < end_date:
    
    
    X_train = X[(X.index >= train_start) & (X.index < test_start)]
    y_train = y[(y.index >= train_start) & (y.index < test_start)]

    X_test = X[(X.index >= test_start) & (X.index <= test_end)]
    y_test = y[(y.index >= test_start) & (y.index <= test_end)]
    
    regr = LinearRegression()
    regr.fit(X_train, y_train)
    regr_score = regr.score(X_train, y_train)   
    
    y_pred = regr.predict(X_test)

    train_start += test_period 
    test_start = train_start + train_period
    test_end = test_start + test_period 

    curr_date = train_start
    ax.plot(y_test.index, y_test, 'b-', label = "Actual Price")
    ax.plot(y_test.index, y_pred, 'r--',  label = "Pred Price")
    




    


        
```
