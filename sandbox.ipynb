```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

pd.set_option('display.mpl_style', 'default')

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
import pandas.io.data as web
```

```python
spc = web.DataReader(name = "^GSPC", data_source = 'yahoo', start = '2010-1-1')
```

```python
# Daily Log returns

spc['Returns'] = np.log(spc['Adj Close'] / spc['Adj Close'].shift(1))    
spc["Returns"].plot()
```

```python
spc['MA42'] = pd.rolling_mean(spc['Adj Close'], window = 42)
spc['MA252'] = pd.rolling_mean(spc['Adj Close'], window = 252)
#spc['STD250'] = pd.rolling_std(spc['Adj Close'], 250, min_periods = 10)
spc[['Adj Close', 'MA252', 'MA42']].plot()
```

```python
import math
# Mov_Vol - Moving volatility
spc['Mov_Vol'] = pd.rolling_std(spc['Returns'], window = 252) * math.sqrt(252)
spc[['Returns', 'Mov_Vol']].plot()
```

```python
train_start = dt.datetime(2014, 4, 1)
test_start = dt.datetime(2014, 8, 1)
```

```python
X_train = spc[['Mov_Vol', "Volume", "Returns"]][(spc.index > train_start) & (spc.index < test_start)]
y_train = spc["Adj Close"][(spc.index > train_start) & (spc.index < test_start)]

# Use to fix "Series" if only one feature is used
#X_train = X_train.reshape(X_train.shape[0], 1)
#y_train = y_train.reshape(y_train.shape[0], 1)
```

```python
X_test = spc[['Mov_Vol', "Volume", "Returns"]][(spc.index >= test_start)]
y_test = spc["Adj Close"][(spc.index >= test_start)]
#X_test = X_test.reshape(X_test.shape[0], 1)
#y_test = y_test.reshape(y_test.shape[0], 1)
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X_train, y_train)

print 'Training score ', regr.score(X_train, y_train)
print 'Testing score ', regr.score(X_test, y_test)
```

```python
y_test.plot()
plt.plot(X_test.index, regr.predict(X_test), color='red')
plt.show()
```

```python
from scipy import stats
import statsmodels.api as sm
```

```python
from statsmodels.graphics.api import qqplot
```

```python
dta = pd.DataFrame(spc["Adj Close"])
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(dta.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(dta, lags=40, ax=ax2)
```

```python
arma_mod30 = sm.tsa.ARMA(dta, (3,0)).fit()
```

```python
print(arma_mod30.params)
print(arma_mod30.aic, arma_mod30.bic, arma_mod30.hqic)
print sm.stats.durbin_watson(arma_mod30.resid.values)
```

```python
resid = arma_mod30.resid
print stats.normaltest(resid)
fig = qqplot(resid, line='q', fit=True)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags=40, ax=ax2)
```

```python
r,q,p = sm.tsa.acf(resid.values.squeeze(), qstat=True)
data = np.c_[range(1,41), r[1:], q, p]
table = pd.DataFrame(data, columns=['lag', "AC", "Q", "Prob(>Q)"])
print(table.set_index('lag'))
```

```python
def mean_forecast_err(y, yhat):
    return y.sub(yhat).mean() 

ranges = [('2010-02', '2011-11', 'r--'), ('2012-04', '2013-01', 'm--'), ('2013-04', '2014-01', 'k--'),('2014-01', '2014-07', 'g--')]
mean_err = []

ax = dta.plot(title = 'Dynamic Prediction')
for r in ranges:
    output = arma_mod30.predict(r[0], r[1], dynamic = True)
    ax = output.plot(ax = ax, style = r[2], linewidth = 3)
    mean_err.append(mean_forecast_err(dta["Adj Close"], output))
plt.show()

print "Mean forecast error for different time ranges"
for i in range(len(mean_err)):
    r = ranges[i]
    print "Period: ", r[0], " - ", r[1], " : ", mean_err[i]
```
