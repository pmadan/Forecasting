```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

pd.set_option('display.mpl_style', 'default')

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
import pandas.io.data as web
```

```python
AAPL = web.DataReader(name = 'AAPL', data_source = 'yahoo', start = '2008-1-1')
```

```python
AAPL["Adj Close"].plot()
```

```python
# Daily Log returns
# Using Adj Close instead of Close, to deal with 7:1 Stock split

AAPL['Returns'] = np.log(AAPL['Adj Close'] / AAPL['Adj Close'].shift(1))    
AAPL["Returns"].plot()
```

```python
AAPL[['Adj Close', 'Returns']].plot(subplots = True)
```

```python
AAPL['42d'] = pd.rolling_mean(AAPL['Adj Close'], window = 42)
AAPL['252d'] = pd.rolling_mean(AAPL['Adj Close'], window = 252)
AAPL[['Adj Close', '42d', '252d']].plot()
```

```python
AAPL['42d'] = pd.rolling_cov(AAPL['Adj Close'], window = 42)
AAPL['252d'] = pd.rolling_cov(AAPL['Adj Close'], window = 252)
AAPL[['Adj Close', '42d', '252d']].plot()
```

```python
import math
# Mov_Vol - Moving volatility
AAPL['Mov_Vol'] = pd.rolling_std(AAPL['Returns'], window = 252) * math.sqrt(252)
AAPL[['Adj Close', 'Mov_Vol', 'Returns']].plot(subplots = True)
```

```python
AAPL.info()
```

```python
X = AAPL.Mov_Vol.dropna()
```

```python
train_start = dt.datetime(2014, 1, 1)
test_start = dt.datetime(2014, 11, 3)
```

```python
X_train = AAPL['Mov_Vol'][(AAPL.index > train_start) & (AAPL.index < test_start)]
y_train = AAPL["Adj Close"][(AAPL.index > train_start) & (AAPL.index < test_start)]
X_train = X_train.reshape(X_train.shape[0], 1)
y_train = y_train.reshape(y_train.shape[0], 1)
```

```python
X_test = AAPL['Mov_Vol'][(AAPL.index >= test_start)]
y_test = AAPL["Adj Close"][(AAPL.index >= test_start)]
X_test = X_test.reshape(X_test.shape[0], 1)
y_test = y_test.reshape(y_test.shape[0], 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X_train, y_train)

print 'Training score ', regr.score(X_train, y_train)
print 'Testing score ', regr.score(X_test, y_test)
```

```python
from __future__ import print_function
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm
```

```python
from statsmodels.graphics.api import qqplot
```

```python
dta = pd.DataFrame(AAPL["Adj Close"])
dta.info()
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(dta.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(dta, lags=40, ax=ax2)
```

```python
arma_mod20 = sm.tsa.ARMA(dta, (2,0)).fit()
print(arma_mod20.params)
```

```python
arma_mod30 = sm.tsa.ARMA(dta, (3,0)).fit()
```

```python
print(arma_mod20.aic, arma_mod20.bic, arma_mod20.hqic)
```

```python
print(arma_mod30.params)
```

```python
print(arma_mod30.aic, arma_mod30.bic, arma_mod30.hqic)
```

```python
sm.stats.durbin_watson(arma_mod30.resid.values)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax = arma_mod30.resid.plot(ax=ax);
```

```python
resid = arma_mod30.resid
```

```python
stats.normaltest(resid)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
fig = qqplot(resid, line='q', ax=ax, fit=True)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags=40, ax=ax2)
```

```python
r,q,p = sm.tsa.acf(resid.values.squeeze(), qstat=True)
data = np.c_[range(1,41), r[1:], q, p]
table = pd.DataFrame(data, columns=['lag', "AC", "Q", "Prob(>Q)"])
print(table.set_index('lag'))
```

```python
predict_output = arma_mod30.predict('2010', '2013', dynamic=True)
print(predict_output)
```

```python
ax = dta.plot()
ax = predict_output.plot(ax=ax, style='r--', label='Dynamic Prediction')
ax.legend()
plt.show()
```

```python
def mean_forecast_err(y, yhat):
    return y.sub(yhat).mean() 
```

```python
mean_err =  mean_forecast_err(dta["Adj Close"], predict_output)
mean_err
```
