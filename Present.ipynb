## How is the data structured?

The data consists of the historical quotes for the SP500 - Open, Close, High, Low, Volume and Adjust close prices and most importantly all indexed by timestamps. This is a Time series -  where the data is observed and measured against time and the individual details eg: High and Low price points, become the cross-section of the data.

Time series data have a natural temporal ordering. This makes time series analysis distinct from cross-sectional studies, in which there is no natural ordering of the observations (src:wikipedia)

##Preprocessing namely what features do I use? 

This is where I spent most of the time, because I had to generate the features necessary to run the regression model. The selected/generated features are:
+ Daily Log returns : Gives us a measure how much the stock changes from day to day. 
+ Trading Volume:  Trading volume indicates how actively is the stock traded and seems indicative of price's moving direction 
+ Short term Moving average: Smoothes out the daily fluctuations, while retaining the trend
+ Volatility: A measure of risk calculated using the weight moving standard deviation of the returns.

### Partitioning the data for a Training and Testing:
The test/train splitting cross-validation approach selects random points from the entire sample to create a training set and testing set. This works well for cross-sectional data but with time series data, we need a temporal split. Meaning, don't use future data to predict something now and avoid causal modeling. (src: Doing data science). 

There are 2 method used: 
+ Simply partitioning the data based on its time index 
+ Accumulate a moving window of training data

##Predictive Modeling using Linear Regression
We are working with continuous data and I have some output already, how do I model my problem? Supervised Learning - Regression techniques to model a continuous but complicated relationship. Here I am using a training period of 2 years to predict the next 5 months. The model uses plain vanilla linear regression and ridge regression. One thing to note, even as I tested the problem with varying time periods is that, the linear regression always scored better on the training set whereas the ridge regression always scored better on the testing set. Still unsure why? but looks like ridge might be the game-day method.

## Windowed Regression
Lets look at many small models instead of a single large model see if that improves the accuracy. The below experiment, uses a moving window of the training data and regression modelling. Setting up a feedback loop. Since we are looking at shorter time frames, I generated 2 newer features :
+ Moving Volatility: Non-weighted scheme, rolling std-dev for a 30 day window
+ 10 day moving average

The loop compares linear regression with ridge regression and uses the model that scores better on the current training window.
The figure shows the plot of the actual price vs the predicted price for a 4 year period. The red band is a 3% confidence interval in the model.
