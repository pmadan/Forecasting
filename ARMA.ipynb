```python
%matplotlib inline

from scipy import stats
import statsmodels.api as sm
from statsmodels.graphics.api import qqplot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

import pandas.io.data as web

pd.set_option('display.mpl_style', 'default')

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
spc = web.DataReader(name = "^GSPC", data_source = 'yahoo', start = '2010-1-1')
dta = pd.DataFrame(spc["Adj Close"])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(dta.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(dta, lags=40, ax=ax2)
```

```python
#arma_mod20 = sm.tsa.ARMA(dta, (2,0)).fit()
#print(arma_mod20.params)
#print(arma_mod20.aic, arma_mod20.bic, arma_mod20.hqic)
# LinAlgError: SVD did not converge
```

```python
arma_mod30 = sm.tsa.ARMA(dta, (3, 0)).fit()
print(arma_mod30.params)
print(arma_mod30.aic, arma_mod30.bic, arma_mod30.hqic)
print sm.stats.durbin_watson(arma_mod30.resid.values)
```

```python
#arma_mod30.resid.plot()
resid = arma_mod30.resid
print stats.normaltest(resid)
fig = qqplot(resid, line = 'q', fit = True)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags = 40, ax = ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags = 40, ax = ax2)
```

```python
r, q, p = sm.tsa.acf(resid.values.squeeze(), qstat = True)
data = np.c_[range(1,41), r[1:], q, p]
table = pd.DataFrame(data, columns = ['lag', "AC", "Q", "Prob(>Q)"])
print(table.set_index('lag'))
```

```python
def mean_forecast_err(y, yhat):
    return y.sub(yhat).mean() 

ranges = [('2010-02', '2011-02', 'r--'), ('2012-04', '2013-01', 'm--'), ('2013-04', '2014-01', 'k--'),('2014-01', '2014-07', 'g--')]
mean_err = []

ax = dta.plot(title = 'Dynamic Prediction')
for r in ranges:
    output = arma_mod30.predict(r[0], r[1], dynamic = True)
    ax = output.plot(ax = ax, style = r[2], linewidth = 3)
    mean_err.append(mean_forecast_err(dta["Adj Close"], output))
plt.show()

print "Mean forecast error for different time ranges"
for i in range(len(mean_err)):
    r = ranges[i]
    print "Period: ", r[0], " - ", r[1], " : ", mean_err[i]
```
